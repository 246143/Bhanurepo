```python
import pandas as pd


titles = pd.read_csv('cast.csv', index_col=None)
titles.head()
```

# 1.What are the ten most common movies names of all time

```python

top_movie_titles = titles['title'].value_counts().head(10)


print(top_movie_titles)
```

# 2. Which three years of the 1930s saw the most films released .

```python

films_in_1930s = titles[(titles['year'] >= 1930) & (titles['year'] <= 1939)]


film_counts_by_year = films_in_1930s['year'].value_counts().reset_index()
film_counts_by_year.columns = ['Year', 'Number of Films']


top_three_years = film_counts_by_year.sort_values(by='Number of Films', ascending=False).head(3)


print(top_three_years)
```

# 5.Plot the number pf "Rustle" characters in each decade of the history of film.

```python
import pandas as pd
cast = pd.read_csv('cast.csv', index_col=None)
cast.head()





import matplotlib.pyplot as plt


rustle_characters = cast[cast['character'] == 'Rustle']


rustle_characters['decade'] = rustle_characters['year'] // 10 * 10


rustle_characters_by_decade = rustle_characters.groupby('decade').size().reset_index(name='count')


plt.figure(figsize=(10, 6))
plt.bar(rustle_characters_by_decade['decade'], rustle_characters_by_decade['count'])
plt.xlabel('Decade')
plt.ylabel('Number of "Rustle" Characters')
plt.title('Number of "Rustle" Characters in Each Decade')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

# 4.plot the number of "Halmet" films made each decade

```python
import pandas as pd
titles = pd.read_csv('cast.csv', index_col=None)
titles.head()




import matplotlib.pyplot as plt

# Filter for films with the title "Hamlet"
hamlet_films = titles[titles['title'] == 'Hamlet']

# Extract the decade from the 'year' column and create a new 'decade' column
hamlet_films['decade'] = hamlet_films['year'] // 10 * 10

# Group the data by decade and count the number of "Hamlet" films in each decade
hamlet_films_by_decade = hamlet_films.groupby('decade').size().reset_index(name='count')

# Create a bar plot to visualize the number of "Hamlet" films in each decade
plt.figure(figsize=(10, 6))
plt.bar(hamlet_films_by_decade['decade'], hamlet_films_by_decade['count'])
plt.xlabel('Decade')
plt.ylabel('Number of "Hamlet" Films')
plt.title('Number of "Hamlet" Films Made Each Decade')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

# 3. Plot the number of films that have been released each decade over the history of cinema

```python
import matplotlib.pyplot as plt


titles['decade'] = titles['year'] // 10 * 10


films_by_decade = titles.groupby('decade').size().reset_index(name='count')


plt.figure(figsize=(12, 6))
plt.bar(films_by_decade['decade'], films_by_decade['count'], width=8)
plt.xlabel('Decade')
plt.ylabel('Number of Films Released')
plt.title('Number of Films Released Each Decade')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

# 13.Plot n-values of the roles that Judi Dench his played over her carrer

```python
import matplotlib.pyplot as plt


judi_dench_roles = cast[cast['name'] == 'Judi Dench']


judi_dench_roles = judi_dench_roles.dropna(subset=['n'])


plt.figure(figsize=(10, 6))
plt.scatter(judi_dench_roles['year'], judi_dench_roles['n'], marker='o', color='b', alpha=0.5)
plt.xlabel('Year')
plt.ylabel('n-value')
plt.title('n-values of Roles Played by Judi Dench Over Her Career')
plt.grid()
plt.show()
```

# 15.How many leading(n=1) roles were available to actors, and how many to actresses, in the 1950s

```python

leading_roles_1950s = cast[(cast['n'] == 1) & (cast['year'] >= 1950) & (cast['year'] <= 1959)]


leading_roles_actors = leading_roles_1950s[leading_roles_1950s['type'] == 'actor']
leading_roles_actresses = leading_roles_1950s[leading_roles_1950s['type'] == 'actress']


num_leading_roles_actors = leading_roles_actors.shape[0]
num_leading_roles_actresses = leading_roles_actresses.shape[0]

print(f"Number of leading roles (n=1) for actors in the 1950s: {num_leading_roles_actors}")
print(f"Number of leading roles (n=1) for actresses in the 1950s: {num_leading_roles_actresses}")
```
