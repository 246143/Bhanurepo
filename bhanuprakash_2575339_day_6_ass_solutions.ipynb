```python
import datetime

class Book:
    def __init__(self, book_id, title, author, publication_date):
        self.__book_id = book_id
        self.__title = title
        self.__author = author
        self.__publication_date = publication_date
        self.__is_checked_out = False

    def get_book_id(self):
        return self.__book_id

    def get_title(self):
        return self.__title

    def get_author(self):
        return self.__author

    def get_publication_date(self):
        return self.__publication_date

    def is_checked_out(self):
        return self.__is_checked_out

    def check_out(self):
        if not self.__is_checked_out:
            self.__is_checked_out = True
            return True
        else:
            return False

    def return_book(self):
        if self.__is_checked_out:
            self.__is_checked_out = False
            return True
        else:
            return False

class Patron:
    def __init__(self, patron_id, name):
        self.__patron_id = patron_id
        self.__name = name

    def get_patron_id(self):
        return self.__patron_id

    def get_name(self):
        return self.__name

class Transaction:
    def __init__(self, book, patron, checkout_date):
        self.__book = book
        self.__patron = patron
        self.__checkout_date = checkout_date
        self.__return_date = None

    def get_book(self):
        return self.__book

    def get_patron(self):
        return self.__patron

    def get_checkout_date(self):
        return self.__checkout_date

    def get_return_date(self):
        return self.__return_date

    def return_book(self):
        self.__return_date = datetime.date.today()
        self.__book.return_book()

    def is_overdue(self):
        if self.__return_date is None:
            return datetime.date.today() > self.__checkout_date
        else:
            return self.__return_date > self.__checkout_date

class FictionBook(Book):
    def __init__(self, book_id, title, author, publication_date, genre):
        super().__init__(book_id, title, author, publication_date)
        self.__genre = genre

    def get_genre(self):
        return self.__genre

class NonFictionBook(Book):
    def __init__(self, book_id, title, author, publication_date, topic):
        super().__init__(book_id, title, author, publication_date)
        self.__topic = topic

    def get_topic(self):
        return self.__topic

class Library:
    def __init__(self):
        self.books = []
        self.patrons = []
        self.transactions = []

    def add_book(self, book):
        self.books.append(book)

    def add_patron(self, patron):
        self.patrons.append(patron)

    def checkout_book(self, book, patron):
        if book.check_out():
            checkout_date = datetime.date.today()
            transaction = Transaction(book, patron, checkout_date)
            self.transactions.append(transaction)
            return True
        else:
            return False

    def return_book(self, transaction):
        transaction.return_book()

    def get_overdue_books(self):
        overdue_books = []
        for transaction in self.transactions:
            if transaction.is_overdue():
                overdue_books.append(transaction.get_book())
        return overdue_books



    library = Library()

    fiction_book = FictionBook(1, "Fiction Book", "Author A", "2022-01-01", "Mystery")
    non_fiction_book = NonFictionBook(2, "Non-Fiction Book", "Author B", "2022-02-01", "Science")

    patron1 = Patron(101, "Patron 1")

    library.add_book(fiction_book)
    library.add_book(non_fiction_book)
    library.add_patron(patron1)

    if library.checkout_book(fiction_book, patron1):
        print("Book checked out successfully.")
    else:
        print("Book is already checked out.")

    library.return_book(library.transactions[0])

    overdue_books = library.get_overdue_books()
    if overdue_books:
        print("Overdue books:")
        for book in overdue_books:
            print(f"- {book.get_title()} by {book.get_author()}")
    else:
        print("No overdue books.")
```

# 7. You are developing an online quiz application where users can take quizzes on various topics and receive scores.

1. Create a class for quizzes and questions.

2. Implement a scoring system that calculates the user's score on a quiz.

3. How would you store and retrieve user progress, including quiz history and scores?

```python
class Question:
    def __init__(self, text, correct_answer):
        self.text = text
        self.correct_answer = correct_answer

class Quiz:
    def __init__(self, name):
        self.name = name
        self.questions = []

    def add_question(self, question_text, correct_answer):
        question = Question(question_text, correct_answer)
        self.questions.append(question)

    def take_quiz(self):
        score = 0
        for question in self.questions:
            print(question.text)
            answer = input("Your answer: ")
            if answer.lower() == question.correct_answer.lower():
                score += 1
        return score

class QuizApp:
    def __init__(self):
        self.users = {}  # Dictionary to store user progress

    def register_user(self, username):
        if username not in self.users:
            self.users[username] = {}

    def take_quiz(self, username, quiz):
        if username in self.users:
            user_progress = self.users[username]
            if quiz.name not in user_progress:
                user_progress[quiz.name] = quiz.take_quiz()
            else:
                print("You've already completed this quiz.")
        else:
            print("User not registered. Please register first.")

    def get_user_progress(self, username):
        if username in self.users:
            return self.users[username]
        else:
            print("User not found.")
            
math_quiz = Quiz("Math Quiz")
math_quiz.add_question("What is 2 + 2?", "4")
math_quiz.add_question("What is 3 x 5?", "15")

science_quiz = Quiz("Science Quiz")
science_quiz.add_question("What is the firstname of megastar?", "chiranjeevi")
science_quiz.add_question("What is the name of ramcharan?", "mega power")

# Create a quiz app
quiz_app = QuizApp()

# Register users
quiz_app.register_user("Alice")
quiz_app.register_user("Bob")

# Users take quizzes
quiz_app.take_quiz("Alice", math_quiz)
quiz_app.take_quiz("Bob", science_quiz)

# Get user progress
alice_progress = quiz_app.get_user_progress("Alice")
bob_progress = quiz_app.get_user_progress("Bob")

print("Alice's Progress:", alice_progress)
print("Bob's Progress:", bob_progress)
```
