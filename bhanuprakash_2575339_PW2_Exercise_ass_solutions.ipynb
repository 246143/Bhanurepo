```python
class Car:
    def __init__(self, manufacturer, car_type, car_age, km_driven, last_service_date, rent_per_day, threshold_distance):
        self.manufacturer = manufacturer
        self.car_type = car_type
        self.car_age = car_age
        self.km_driven = km_driven
        self.last_service_date = last_service_date
        self.rent_per_day = rent_per_day
        self.available = True
        self.threshold_distance = threshold_distance
        self.service_history = []

    def service(self, notes):
        today = datetime.today().date()
        self.service_history.append({"date": today, "notes": notes})
        self.last_service_date = today
        self.km_driven = 0  # Reset the driven distance after servicing

    def display(self):
        return (f"{self.manufacturer}, {self.car_type}, {self.car_age} years old, "
                f"{self.km_driven} km driven, Last serviced: {self.last_service_date}, Available: {self.available}")


class RentalSystem:
    def __init__(self):
        self.cars = []
        self.total_earnings = 0

    def list_available_cars(self):
        return [car for car in self.cars if car.available]

    def rent_out_car(self, car, days):
        if car.available:
            car.available = False
            rent = days * car.rent_per_day
            self.total_earnings += rent
            return rent
        else:
            return None

    def add_car(self, car):
        self.cars.append(car)

    def remove_car(self, car):
        if car in self.cars:
            self.cars.remove(car)


class Admin:
    def __init__(self, system):
        self.system = system

    def add_new_car(self, manufacturer, car_type, car_age, km_driven, last_service_date, rent_per_day, threshold_distance):
        new_car = Car(manufacturer, car_type, car_age, km_driven, last_service_date, rent_per_day, threshold_distance)
        self.system.add_car(new_car)

    def remove_car(self, car):
        self.system.remove_car(car)

    def service_car(self, car, notes):
        car.service(notes)

    def check_service(self):
        for car in self.system.cars:
            if car.km_driven >= car.threshold_distance:
                print(f"{car.manufacturer} needs service!")


class Customer:
    def __init__(self, name):
        self.name = name

    def view_cars(self, system):
        return system.list_available_cars()

    def rent_car(self, system, car, days):
        cost = system.rent_out_car(car, days)
        if cost:
            return f"{self.name} has rented the {car.manufacturer} for {days} days. Total cost: ${cost}"
        else:
            return "Car not available for rent."

    def return_car(self, car):
        car.available = True
        print(f"{self.name} has returned the {car.manufacturer}.")


# Sample Usage:
system = RentalSystem()

admin = Admin(system)
admin.add_new_car("Toyota", "Basic", 3, 15000, "2020-05-15", 50, 20000)
admin.add_new_car("BMW", "Luxury", 1, 5000, "2021-02-10", 200, 15000)
admin.add_new_car("Hyundai", "Mid-Range", 2, 11000, "2020-08-20", 100, 19000)

customer = Customer("John")

cars = customer.view_cars(system)
print(cars[0].display())
print(customer.rent_car(system, cars[0], 5))

admin.check_service()
```

# 2.Railway Management System

Problem Statement:

The problem statement is to create the Railway Management System that develops a user-friendly software application that facilitates various functionalities related to railway ticketing and management. The system should allow users to book tickets, cancel booking, check fares. view their bookings, and display available trains.

• Case Study:

Imagine a scenario where a railway company needs a modern and efficient system to handle their ticketing and management operations. The current manual process is time-consuming, prone to errors, and lacks flexibility. They want to automate these processes and provide a convenient platform for their customers to book tickets, check fares, and manage their reservations easily.

• Data Analysis Part:

The data analysis part of the Railway Management System involves analysing the data collected from various sources, such as customer bookings, train schedules, and fare information. This analysis helps in gaining insights into customer preferences, popular routes etc.
Objective:

The primary objective of the Railway Management System is to streamline and automate the ticketing and management processes for both the railway company and its customers. The specific objectives include:

✓ Ticket Booking: Allow users to book tickets conveniently by selecting the desired train, class, and travel date. The system should validate the availability of seats and provide a confirmation for successful bookings.

✓ Reservation Cancellation: Provide users with the option to cancel their booked tickets if required. The system should handle cancellation requests, update seat availability, and process refunds if applicable.

✓ Fare Checking: Enable users to check fares for different trains, classes, and routes before making a booking. The system should fetch the fare details from the database and display them accurately.

✓ Booking History: Allow users to view their booking history, including details such as train name, departure/arrival times, class, and booking status.
This feature helps users keep track of their travel history.

✓ Train Availability: Provide users with an up-to-date list of available trains based on the selected source and destination. The system should display relevant details like train name, departure/arrival times, and seat availability.

✓Menu: Display a menu to users, showcasing the available options and functionalities of the system.

✓ Exit: Allow users to exit the program gracefully when they have completed their tasks.

```python
class Train:
    def __init__(self, train_id, train_name, source, destination, departure_time, arrival_time):
        self.train_id = train_id
        self.train_name = train_name
        self.source = source
        self.destination = destination
        self.departure_time = departure_time
        self.arrival_time = arrival_time
        self.seat_availability = {'1st_class': 10, '2nd_class': 20}
        self.fare = {'1st_class': 1000, '2nd_class': 500}

    def show_train_info(self):
        return f"{self.train_name} ({self.train_id}) - {self.source} to {self.destination} - Dep: {self.departure_time} Arr: {self.arrival_time}"

    def check_availability(self, class_type):
        return self.seat_availability[class_type]

    def book_seats(self, class_type, num_seats):
        if self.seat_availability[class_type] >= num_seats:
            self.seat_availability[class_type] -= num_seats
            return True
        else:
            return False

    def cancel_seats(self, class_type, num_seats):
        self.seat_availability[class_type] += num_seats


class Ticket:
    ticket_counter = 0

    def __init__(self, user, train, class_type):
        Ticket.ticket_counter += 1
        self.ticket_id = Ticket.ticket_counter
        self.user = user
        self.train = train
        self.class_type = class_type
        self.fare = train.fare[class_type]
        self.booking_date = "today"  # use datetime in real scenarios

    def show_ticket_info(self):
        return f"Ticket ID: {self.ticket_id} | Train: {self.train.show_train_info()} | Booked for: {self.user.user_name} in {self.class_type} | Fare: {self.fare}"


class User:
    def __init__(self, user_name):
        self.user_name = user_name
        self.bookings = []

    def book_ticket(self, train, class_type):
        if train.book_seats(class_type, 1):
            ticket = Ticket(self, train, class_type)
            self.bookings.append(ticket)
            return ticket
        else:
            return None

    def view_history(self):
        for ticket in self.bookings:
            print(ticket.show_ticket_info())


class ReservationSystem:
    def __init__(self):
        self.trains = [
            Train(1, "Express", "A", "B", "09:00", "12:00"),
            Train(2, "FastTrack", "A", "C", "10:00", "13:00")
        ]
        self.users = []

    def register_user(self, user_name):
        user = User(user_name)
        self.users.append(user)
        return user

    def search_trains(self, source, destination):
        available_trains = [train for train in self.trains if train.source == source and train.destination == destination]
        return available_trains


# Main Program
def main():
    system = ReservationSystem()
    user1 = system.register_user("John")
    user2 = system.register_user("Jane")

    # John wants to view available trains from A to B
    trains = system.search_trains("A", "B")
    for train in trains:
        print(train.show_train_info())

    # John books a ticket
    ticket = user1.book_ticket(trains[0], '1st_class')
    if ticket:
        print("Ticket booked successfully!")
        print(ticket.show_ticket_info())
    else:
        print("Booking failed!")

    # Jane books a ticket in the same train
    ticket = user2.book_ticket(trains[0], '1st_class')
    if ticket:
        print("Ticket booked successfully!")
        print(ticket.show_ticket_info())
    else:
        print("Booking failed!")

    # John views his booking history
    user1.view_history()

```
